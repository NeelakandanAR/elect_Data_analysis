```python


#Loading libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Loading datasets

elections = pd.read_csv('/content/governors_county_candidate.csv')
```

```python


#Dropping unnecessary variables in elections dataset
elections.drop(['state','county', 'candidate'],axis = 'columns',inplace = True)
elections.head()

```

#party who won this year

```python
winner = elections.groupby(['party', 'won']).size()
winner = winner.to_frame()
winner = winner.reset_index()

winner.sort_values(by='won', ascending=False)

#clearly the republican party won the election with majority
```

Total number of votes by the party

```python
#party who won this year
this_year = elections.groupby(['party'])['votes'].sum()
this_year = this_year.to_frame()
this_year = this_year.reset_index()

# winner = election_winner.sort_values(by='votes', ascending = False)
# df.sort_values(by='col1', ascending=False)
```

```python
this_year.sort_values(by='votes', ascending=False)
# The republican party won the election this year
```

```python
# df = this_year.DataFrame()
# ax = this_year.plot.bar(x='party', y='votes', rot=0)
this_year.sort_values('votes', ascending=False)[['party','votes']].plot.bar(stacked=True)
```

percentage of the votes by the party

```python
#which party got the votes in percentage
# Calculating Percentage
this_year['percent'] = (this_year['votes'] / 
                  this_year['votes'].sum()) * 100
this_year.sort_values(by='percent', ascending=False)
```
